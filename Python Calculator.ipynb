```python
# Project Requirements
#•	Basic operations: +, -, *, /
#•	Advanced: %, ** (power), √ (square root)
#•	History: remember last 5 calculations
#•	Error handling: division by zero, invalid input
#•	Clean menu interface
history=[]
def calculate(a,op,b):
  if(op=='+'):
    return a+b
  elif(op=='-'):
    return a-b
  elif(op=='*'):
    return a*b
  elif(op=='/'):
    if b==0:
      raise ZeroDivisionError(" Cannot divide by zero")
    return a/b
  else:
    raise ValueError("Unknown operand")


def main():
  print('-'*45)
  print("               Python Calculator")
  print('-'*45)

  while True:
    print('Options: c for calculate, h for history, q for quit')
    choice= input('choose: ')

    if choice=='q':
      break
    elif choice=='c':
      num1= int(input('Enter first number'))
      op= input('choose operation +,-,*,/')
      num2=int(input('Enter second number'))
      result=calculate(num1, op, num2)
      entry=f"{num1}{op}{num2}={result}"
      history.append(entry)
      print(result)

    elif choice=='h':
      if not history:
        print("No history yet")
      else:
        for h in history[-5:]:
          print(f" {h}")
# calling function
main()
```
